```python
text='''Around 65% (2.37 lakh) of all road accidents in India 
took place on straight roads in 2020, government data has shown. 
Such accidents killed over 85,000 people, 
the 'Road accidents in India- 2020' report released by the government this year, said. 
Over 72% of accidents and 67% of fatalities took place under sunny or clear weather in 2020, it added. '''
```

```python
summarizer = pipeline("summarization")
result = summarizer(text, max_length=130, min_length=30,
do_sample=False)
print("Original Article")
print('*' * 100)
pprint.pprint(text)
print('*' * 100)
print("\n")
print("Summary")
print('*' * 100)
pprint.pprint(result[0]['summary_text'])
print('*' * 100)
```

```python
from transformers import AutoTokenizer, AutoModel, BartForConditionalGeneration

model_name = "sshleifer/distilbart-cnn-12-6"
model = BartForConditionalGeneration.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

summarizer2 = pipeline("summarization", model=model, tokenizer=tokenizer)
result = summarizer2(text, max_length=130, min_length=30, do_sample=False)
pprint.pprint(result[0]['summary_text'])
```

```python
save_dir = "saved"
tokenizer.save_pretrained(save_dir)
model.save_pretrained(save_dir)

tok = AutoTokenizer.from_pretrained(save_dir)
model = AutoModel.from_pretrained(save_dir)
```
